```python
%sql
CREATE OR REPLACE VIEW ecommerce.gold.top_products AS
SELECT
  product_id,
  views,
  purchases,
  revenue,
  conversion_rate
FROM ecommerce.gold.products
WHERE purchases > 10
ORDER BY revenue DESC
LIMIT 100;
```
