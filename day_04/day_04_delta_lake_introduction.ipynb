```python
path_oct = "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv"
events = spark.read.csv(path_oct, header=True, inferSchema=True)
```

```python
# Convert to Delta
events.write.format("delta").mode("overwrite").save("/Volumes/workspace/ecommerce/ecommerce_data/delta/events")
```

```python

# Create managed table
events.write.format("delta").saveAsTable("events_table")
```

```python
# SQL approach
spark.sql("""
    CREATE TABLE events_delta
    USING DELTA
    AS SELECT * FROM events_table
""")
```

```python
# Test schema enforcement
try:
    wrong_schema = spark.createDataFrame([("a","b","c")], ["x","y","z"])
    wrong_schema.write.format("delta").mode("append").save("/delta/events")
except Exception as e:
    print(f"Schema enforcement: {e}")
```
