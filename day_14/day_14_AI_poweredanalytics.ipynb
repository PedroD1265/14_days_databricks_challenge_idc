```python
%sql
SELECT
  category_code,
  ROUND(SUM(price), 2) AS revenue
FROM ecommerce.silver.events
WHERE event_type = 'purchase'
  AND category_code IS NOT NULL
GROUP BY category_code
ORDER BY revenue DESC;
```

```python
%sql
SELECT
  product_id, views, purchases, revenue, conversion_rate
FROM ecommerce.gold.products
WHERE views > 0
ORDER BY conversion_rate DESC
LIMIT 20;
```

```python
%sql
SELECT
  event_date,
  COUNT(*) AS purchases
FROM ecommerce.silver.events
WHERE event_type = 'purchase'
GROUP BY event_date
ORDER BY event_date;
```

```python
%sql
WITH viewed AS (
  SELECT DISTINCT user_id
  FROM ecommerce.silver.events
  WHERE event_type = 'view'
),
purchased AS (
  SELECT DISTINCT user_id
  FROM ecommerce.silver.events
  WHERE event_type = 'purchase'
)
SELECT v.user_id
FROM viewed v
LEFT ANTI JOIN purchased p
ON v.user_id = p.user_id
LIMIT 100;
```

```python
from transformers import pipeline
import mlflow

classifier = pipeline("sentiment-analysis")  # descargarÃ¡ modelo la primera vez

reviews = [
    "This product is amazing!",
    "Terrible quality, waste of money"
]
results = classifier(reviews)

with mlflow.start_run(run_name="sentiment_baseline"):
    mlflow.log_param("task", "sentiment-analysis")
    mlflow.log_param("model", "transformers-pipeline-default")
    # ejemplo simple: guardamos #positivos
    positives = sum(1 for r in results if r["label"].upper().startswith("POS"))
    mlflow.log_metric("positives", positives)

results
```
